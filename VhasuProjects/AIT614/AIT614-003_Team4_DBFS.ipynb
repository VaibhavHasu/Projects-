```

# Uploading dataset
df_clean = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/shared_uploads/akolagan@gmu.edu/FHWA_Bridge_Conditions_Dataset.csv")

# Generating summary statistics and written directly to DBFS as CSV
summary_df = df_clean.describe()
summary_df.write.mode("overwrite").csv(
    "dbfs:/FileStore/summary_stats.csv",
    header=True
)

# Listing the contents of the new summary_stats directory
files = dbutils.fs.ls("dbfs:/FileStore/summary_stats.csv")
display(files)

# Read back the first 200 bytes of the first part file to validate its contents
first_part = files[0].path
print(dbutils.fs.head(first_part, 200))

# Copying that CSV folder into an archive directory
dbutils.fs.mkdirs("dbfs:/FileStore/archive")
dbutils.fs.cp(
    "dbfs:/FileStore/summary_stats.csv",
    "dbfs:/FileStore/archive/summary_stats_backup.csv",
    recurse=True
)

# Removing the original summary_stats folder to clean up
dbutils.fs.rm("dbfs:/FileStore/summary_stats.csv", recurse=True)
```
