**Importing Libraries**

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

**Loading and cleaning the datasett**

```python
# Load and clean the dataset
df = pd.read_csv("/FHWA_Bridge_Conditions_Dataset.csv")

# Convert key columns to numeric
df["Age_Years"] = pd.to_numeric(df["Age_Years"], errors='coerce')
df["Deck_Condition"] = pd.to_numeric(df["Deck_Condition"], errors='coerce')
df["Daily_Traffic"] = pd.to_numeric(df["Daily_Traffic"], errors='coerce')
df["Truck_Percentage"] = pd.to_numeric(df["Truck_Percentage"], errors='coerce')
df["Repair_Cost_USD"] = df["Repair_Cost_USD"].replace('[\$,]', '', regex=True).astype(float)

# Drop rows with missing key values
df = df.dropna(subset=["Age_Years", "Deck_Condition", "Daily_Traffic", "Repair_Cost_USD", "Material"])
```

```python
# Set style
sns.set(style="whitegrid")
```

**Plot 1: Deck Condition Distribution**

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Deck_Condition"], bins=10, kde=True, color='skyblue')
plt.title("Distribution of Bridge Deck Condition")
plt.xlabel("Deck Condition")
plt.ylabel("Count")
plt.show()
```

**Plot 2: Plot of Repair Cost vs Daily Traffic**

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="Daily_Traffic", y="Repair_Cost_USD", data=df, alpha=0.3)
plt.title("Repair Cost vs. Daily Traffic")
plt.xlabel("Daily Traffic")
plt.ylabel("Repair Cost (USD)")
plt.xscale("log")
plt.yscale("log")
plt.grid(True, which="both", linestyle='--', linewidth=0.5)
plt.show()
```

**Plot 3: Plot of Average Deck Condition by Material**

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Material", y="Deck_Condition", data=df, estimator=np.mean, ci=None)
plt.title("Average Deck Condition by Material")
plt.xticks(rotation=45)
plt.ylabel("Average Deck Condition")
plt.show()
```
